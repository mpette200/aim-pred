```python
import json
import time
import requests
import datetime as dt
import numpy as np
import pandas as pd
from urllib import request, parse
```

```python
tickers = [
    'PAF','TM17','IOM','YNGN','ESL','ASC','MIDW','CVSG','GBG','IPX',
    'BREE','MAB1','KAPE','HGM','ATYM','TFW','DTG','BMN','PRSM','WAND',
    'LTG','HUR','SIR','DATA','JSG','NUM','SUMO','BRK','ABC','KWS',
    'TRMR','GTLY','SQZ','AFX','TCM','POLR','BVXP','KGH','KETL','DOTD',
    'PURP','IGR','BUR','BOO','TUNE','TRCS','SCPA','RQIH','PEBB','MRL',
    'LGRS','JHD','HCM','GHH','CREO','BMK','ASY','APGN','AFM','FEVR',
    'GRP','SLN','CTH','HZD','RWS','RST','CRW','AMS','HOTC','FDEV','GAMA',
    'APH','NFC','MPE','CHRT','PMI','CAM','ITM','GROW','EMIS','YNGA',
    'CAML','MTW','YOU','NICL','RNWH','CLIN','JSE','IMO','WJG','WHR',
    'SMS','CWR','CDM','ABDP','IQE','FDP','VCP','UPR','IDEA']
```

```python
# 'https://query1.finance.yahoo.com/v7/finance/download/{}?period1=1451606400&period2=1483228800&interval=1d&events=history'
url_prices = 'https://query1.finance.yahoo.com/v7/finance/download/{}?'
start = dt.datetime(2017, 1, 1, 1)
end = dt.datetime(2099, 12, 31, 1)
params_prices = {
    'period1': format(start.timestamp(), '.0f'),
    'period2': format(end.timestamp(), '.0f'),
    'interval': '1d',
    'events': 'history',
}
df_list = []
for ticker in tickers:
    ticker_L = ticker + '.L'
    url = url_prices.format(ticker_L) \
          + parse.urlencode(params_prices)
    print('Fetching URL:  ', url)
    df = pd.read_csv(url)
    df['symbol'] = ticker
    df_list.append(df)
all_prices = pd.concat(df_list, ignore_index=True)
```

```python
# https://query1.finance.yahoo.com/v10/finance/quoteSummary/YOU.L?formatted=true&crumb=0CD5uFjQiFE&lang=en-GB&region=GB&modules=summaryProfile%2CfinancialData%2CrecommendationTrend%2CupgradeDowngradeHistory%2Cearnings%2CdefaultKeyStatistics%2CcalendarEvents%2CesgScores%2Cdetails&corsDomain=uk.finance.yahoo.com
# https://query1.finance.yahoo.com/v10/finance/quoteSummary/YOU.L?formatted=true&lang=en-GB&region=GB&modules=summaryProfile&corsDomain=uk.finance.yahoo.com
url_info = 'https://query1.finance.yahoo.com/v10/finance/quoteSummary/{}?'
params_info = {
    'formatted': 'true',
    'lang': 'en-GB',
    'region': 'GB',
    'modules': 'summaryProfile',
    'corsDomain': 'uk.finance.yahoo.com'
}
prefix = params_info['modules']
drop_prefix = lambda x: x.replace(prefix + '.', '')
df_list = []
for ticker in tickers:
    ticker_L = ticker + '.L'
    url = url_info.format(ticker_L) \
              + parse.urlencode(params_info)
    print('Fetching URL:  ', url)
    df1 = pd.read_json(url)
    df2 = pd.json_normalize(df1.loc['result', 'quoteSummary'])
    df3 = df2.rename(columns=drop_prefix)
    df3['symbol'] = ticker
    df_list.append(df3)
    time.sleep(0.1) # pause to avoid hitting server
all_info = pd.concat(df_list, ignore_index=True)
```
